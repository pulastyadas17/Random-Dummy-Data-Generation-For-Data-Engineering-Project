```python
import random
from faker import Faker
import pandas as pd


fake = Faker()

cities = [
    'New York City', 
    'Los Angeles',
    'Chicago',
    'Houston',
    'Phoenix',
    'Philadelphia',
    'San Antonio',
    'San Diego',
    'Dallas',
    'San Jose'
]


def generate_customer_data(num_customers=100):
    customer_data = []
    for i in range(1,num_customers+1):
        customer_id = f"U{str(i).zfill(3)}"  #Generate CustomerID from U001 to U099 as per fact table
        name = fake.name()
        age = random.randint(18, 90)
        gender = random.choice(["Male", "Female"])
        address = fake.address().replace("\n", ", ")
        city = random.choice(cities)
        phone_number = fake.unique.numerify('##########')
        email_id = fake.email()
        dob = fake.date_of_birth(minimum_age=18,maximum_age=90)
        customer_data.append([customer_id,name,age,gender,address,city,phone_number,email_id, dob])
    
    customer_df = pd.DataFrame(customer_data,columns=['CustomerID','Name','CustomerAge','CustomerGender','FullAddress','City','PhoneNumber','EmailID','DOB'])
    return customer_df

customer_df=generate_customer_data()
```
